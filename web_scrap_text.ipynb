```python
from trafilatura import fetch_url, extract, html2txt
```

```python

url = "https://telnyx.com/resources/gpu-architecture-ai"
```

```python
downloaded = fetch_url(url)
print("RAW HTML\n", downloaded)
```

```python
all_text = html2txt(downloaded)
print("\nALL TEXT\n", all_text)
```

```python

main_text = extract(downloaded)
print("\nMAIN TEXT\n", main_text)
```

```python
import textwrap
```

```python
def chunk_text(text, max_tokens=256):
    # Approx: 1 token ≈ 0.75 word ≈ 4 characters
    max_chars = max_tokens * 4
    return textwrap.wrap(text, width=max_chars)
```

```python

chunks = chunk_text(main_text)
print(f"Total Chunks: {len(chunks)}")
```

```python
chunks
```

```python
dataset = []
```

```python
for idx, chunk in enumerate(chunks):
    entry = {
        "text": f"<s>[INST] Summarize this passage from an AWS ML blog:\n\n{chunk} [/INST]\n\nSure! Here's the summary: "
    }
    dataset.append(entry)
```

```python
dataset
```

```python
with open("chunks.json", "w") as f:
    json.dump(dataset, f, indent=2)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, TrainingArguments, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from trl import SFTTrainer
```

```python
from huggingface_hub import notebook_login
notebook_login()
```

```python
model_name = "meta-llama/Llama-2-7b-hf"
```

```python
# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=False)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"
```

```python
from datasets import Dataset
```

```python
dataset = Dataset.from_list(chunked_dataset)
```

```python
# Tokenize
def tokenize(sample):
    return tokenizer(sample["text"], truncation=True, padding="max_length", max_length=512)
tokenized_dataset = dataset.map(tokenize, batched=True)
```

```python
import os
import torch
from datasets import load_dataset
from transformers import(
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    HfArgumentParser,
    TrainingArguments,
    pipeline,
    logging,
)
from peft import LoraConfig, PeftModel
from trl import SFTTrainer
```

```python
# Quantization config
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype="float16",
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True
)
```

```python
# Load model in 4-bit and apply QLoRA
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    quantization_config=bnb_config
)
```

```python
# Apply LoRA
lora_config = LoraConfig(
    r=8,
    lora_alpha=32,
    lora_dropout=0.05,
    target_modules=["q_proj", "v_proj"],
    bias="none",
    task_type="CAUSAL_LM"
)
```

```python
model = get_peft_model(model, lora_config)
```

```python
training_args = TrainingArguments(
    output_dir="./blog_qlora_finetuned",
    per_device_train_batch_size=2,
    num_train_epochs=3,
    learning_rate=2e-4,
    fp16=True,
    logging_steps=5,
    save_steps=10,
    save_total_limit=1,
    report_to="tensorboard",
)
```

```python
trainer = SFTTrainer(
    model=model,
    train_dataset=tokenized_dataset,
    args=training_args,


)
```

```python
trainer.train()
```
